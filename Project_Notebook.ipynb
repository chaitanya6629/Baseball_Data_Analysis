```python
def max_occurences(x):
    maximum = 0
    for ind, grp in x:
        l = len(grp)
        if l>maximum:
            maximum = l
            pl_id = ind
    return pl_id, maximum
```

```python
x= max_occurences(grouped_by_player)
print x
```
